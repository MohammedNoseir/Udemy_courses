```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import neattext as ntx
```

```python
df=pd.read_csv('udemy_courses_cleaned.csv')
df.head()
```

```python
df['course_title_cleaned']=df['course_title'].apply(ntx.remove_stopwords)
```

```python
df['course_title_cleaned']=df['course_title'].apply(ntx.remove_special_characters)
```

```python
df.head()
```

## the vectorization 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
model=CountVectorizer()
```

```python
df['course_title_cleaned'].duplicated().sum()
```

```python
matrix=model.fit_transform(df['course_title_cleaned']).toarray()
```

```python
matrix.shape
```

## Bulding Recommender System

```python
sim_matrix=cosine_similarity(matrix)
```

```python
def my_rec_system(choose_title):
  course_index=pd.Series(df.index,index=df['course_title'])
  title=choose_title
  course_index[title]
  Scores=list(enumerate(sim_matrix[course_index[title]]))
  sorted_course=sorted(Scores,key=lambda x: x[1] ,reverse=True)
  selected_index=[i[0] for i in sorted_course[1:]]
  selected_scores=[i[1] for i in sorted_course[1:]]
  rec_courses=df.iloc[selected_index]['course_title'].head().values
  return rec_courses
```

```python
my_rec_system('Beginner to Pro - Financial Analysis in Excel 2017')
```
